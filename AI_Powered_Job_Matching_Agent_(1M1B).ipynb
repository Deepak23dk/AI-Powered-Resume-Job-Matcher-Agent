2. IMPORT LIBRARIES


```python
import pdfplumber
import re
import nltk
from google.colab import files
```

3. UPLOAD RESUME


```python
uploaded = files.upload()

```

4. EXTRACT RESUME TEXT

```python
def extract_resume_text(pdf_file):
    text = ""
    with pdfplumber.open(pdf_file) as pdf:
        for page in pdf.pages:
            text += page.extract_text()
    return text

resume_text = extract_resume_text(list(uploaded.keys())[0])
print(resume_text[:1000])
```

5. CLEAN TEXT

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z ]', ' ', text)
    return text

resume_clean = clean_text(resume_text)
```

6. PASTE JOB DESCRIPTION


```python
job_description = """
We are looking for a Software Developer who can design, develop, and maintain software applications.
The candidate should have strong programming and problem-solving skills.

Required Skills:
- Python
- Java
- SQL
- Data Structures and Algorithms
- Object-Oriented Programming
- HTML, CSS, JavaScript
- Git and version control
- Problem solving

Preferred Skills:
- React
- Flask or Django
- Basic knowledge of machine learning
- Team collaboration skills
"""

job_clean = clean_text(job_description)
```

7. SKILL LIST

```python
skills = [
    "python","java","sql","html","css","javascript",
    "react","machine learning","data analysis",
    "flask","django","git","problem solving"
]
```

8. EXTRACT SKILLS


```python
resume_skills = [skill for skill in skills if skill in resume_clean]
jd_skills = [skill for skill in skills if skill in job_clean]

resume_skills, jd_skills
```

9. MATCHING LOGIC


```python
matched_skills = set(resume_skills).intersection(set(jd_skills))
missing_skills = set(jd_skills) - set(resume_skills)

if len(jd_skills) > 0:
    match_percentage = (len(matched_skills) / len(jd_skills)) * 100
else:
    match_percentage = 0

matched_skills, missing_skills, match_percentage
```

10. FINAL RESULT


```python
print("JOB MATCH RESULT")
print("----------------")
print(f"Match Percentage: {round(match_percentage,2)}%")
print("Matched Skills:", matched_skills)
print("Missing Skills:", missing_skills)
```
